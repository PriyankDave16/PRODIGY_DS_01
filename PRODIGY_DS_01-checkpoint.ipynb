```python
df.drop(['Country Name','Indicator Name'],axis=1,inplace=True)
```

# Visualization of Top 10 country by population between 2020 & 2024

```python
total_population=df[df['Indicator Code']=='SP.POP.TOTL']
total_population_sorted=total_population.sort_values(by="2020",ascending=False)  
total_top_ten_countries=total_population_sorted.head(10)
print("top ten countries of total population\n")
print(total_top_ten_countries['Country Code'])
```

```python
plt.figure(figsize =(20,6))
plt.subplot(2,2,1)
sns.barplot(x="2020",y="Country Code",data=total_top_ten_countries,palette="pastel")
plt.title("top ten countries by population(2025)",fontsize=10)
plt.title("total population",fontsize=10)
plt.ylabel("country",fontsize=10)
plt.show()

plt.figure(figsize =(20,6))
plt.subplot(2,2,2)
sns.barplot(x="2024",y="Country Code",data=total_top_ten_countries,palette="pastel")
plt.title("top ten countries by population(2025)",fontsize=10)
plt.title("total population",fontsize=10)
plt.ylabel("country",fontsize=10)
plt.show()
```

# Visualization of bottom ten countries Between 2020 & 2024

```python
total_population=df[df['Indicator Code']=='SP.POP.TOTL']
total_population_sorted=total_population.sort_values(by="2024",ascending=True)  
total_top_ten_countries=total_population_sorted.head(10)
print("Bottom ten countries of total population\n")
print(total_top_ten_countries['Country Code'])
```

```python
plt.figure(figsize =(20,6))
plt.subplot(2,2,1)
sns.barplot(x="2020",y="Country Code",data=total_top_ten_countries,palette="pastel")
plt.title("bottom ten countries by population(2024)",fontsize=10)
plt.title("total population",fontsize=10)
plt.ylabel("country",fontsize=10)
plt.show()

plt.figure(figsize =(20,6))
plt.subplot(2,2,2)
sns.barplot(x="2024",y="Country Code",data=total_top_ten_countries,palette="pastel")
plt.title("bottom ten countries by population(2024)",fontsize=10)
plt.title("total population",fontsize=10)
plt.ylabel("country",fontsize=10)
plt.show()
```
